# Lab 4 - Pretrained Networks: YOLO

```python
from ultralytics import YOLO
from PIL import Image
```

```python
# Load a model
model = YOLO("yolov8s.pt")
names = model.names
```

```python
photo_path = "./lab4_photo.jpg"

im = Image.open(photo_path)
display(im)
```

```python
results = model(photo_path)  # predict on an image
```

```python
for r in results:
    for c in r.boxes.cls:
        print(names[int(c)])
```
